```python
def classified(list1):
    if (list1 == 'Open' or list1 == 'Pending'):
        return 'Open'
    elif (list1 =='Closed' or list1 =='Solved' ):
        return 'Closed'
```

```python
list2 = df['Status']
print (list2)
```

```python
df['Status_new'] = list(map(classified,list2))
```

```python
df
```

## Status_new is the new categorical variable with value as Open and Closed. Open & Pending is categorized as Open and  Closed & Solved is categorized as Closed.
